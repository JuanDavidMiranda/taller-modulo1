### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
# Promedio
suma = 0
contador = 0

for valor in esperanzas_de_vida:
    suma += valor
    contador += 1

promedio = suma / contador
# Máximo
maximo = esperanzas_de_vida[0]

for valor in esperanzas_de_vida:
    if valor > maximo:
        maximo = valor
# Mínimo
minimo = esperanzas_de_vida[0]

for valor in esperanzas_de_vida:
    if valor < minimo:
        minimo = valor

print("Promedio:", promedio)
print("Máximo:", maximo)
print("Mínimo:", minimo)
````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

2. Imprimir cada par: `(valor, clasificación)` usando un bucle `for`

````python
# Tu código aquí
esperanzas_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]

def clasificar_esperanza_vida(valor):
    if valor >= 75:
        return "Alta"
    elif valor >= 60:
        return "Media"
    else:
        return "Baja"

clasificaciones = []

for valor in esperanzas_vida:
    clasificaciones.append(clasificar_esperanza_vida(valor))

for i in range(len(esperanzas_vida)):
    print("(", esperanzas_vida[i], ",", clasificaciones[i], ")")
````

### 2.2 — Limpieza de datos

🔧 **Ejercicio 2.2** — El dataset tiene algunos valores nulos (¡datos reales siempre tienen este problema!).  
1. Rellena los valores nulos de columnas **numéricas** con la **media** de cada columna
2. Verifica que ya no queden nulos
3. Explica en una celda de texto: ¿Por qué usamos la media y no simplemente eliminamos esas filas?

````python
# Tu código aquí
import pandas as pd

df = pd.read_csv("salud_mundial.csv")

columnas_numericas = df.select_dtypes(include=['float64', 'int64']).columns

for col in columnas_numericas:
    media = df[col].mean()
    df[col].fillna(media, inplace=True)

print("Valores nulos por columna después de la limpieza:")
print(df.isnull().sum())
````

### 2.3 — Estadísticas descriptivas

🔧 **Ejercicio 2.3**
1. Muestra el resumen estadístico completo del dataset
2. ¿Cuál es la esperanza de vida **promedio** mundial?
3. ¿Cuál es el país con **mayor** gasto en salud? ¿Y el de **menor**?
4. ¿Cuál es la **mediana** de la mortalidad infantil?

````python
# Tu código aquí
import pandas as pd

df = pd.read_csv("salud_mundial.csv")

print("Resumen estadístico:")
print(df.describe(include='all'))

promedio_vida = df["esperanza_vida"].mean()
print("\nEsperanza de vida promedio mundial:", promedio_vida)

pais_max_gasto = df.loc[df["gasto_salud_usd"].idxmax()]
pais_min_gasto = df.loc[df["gasto_salud_usd"].idxmin()]

print("\nPaís con mayor gasto en salud:")
print(pais_max_gasto["pais"], "-", pais_max_gasto["gasto_salud_usd"])

print("\nPaís con menor gasto en salud:")

print(pais_min_gasto["pais"], "-", pais_min_gasto["gasto_salud_usd"])

mediana_mortalidad = df["mortalidad_infantil"].median()
print("\nMediana de la mortalidad infantil:", mediana_mortalidad)
````

### 2.4 — Filtrado y selección
